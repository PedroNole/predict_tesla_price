```python
import pandas as pd

df_1 = pd.read_csv('df_tesla_final.csv')
df_2 = pd.read_csv('tesla_coches_combinados.csv')
df_1['País'] = 'España'
df_1['price'] = round(df_1['price']*1000,2)
```

```python
def clean_model(row):
    if 'Model S' in row:
        return 'Model S'
    elif 'Model X' in row:
        return 'Model x'
    elif 'Model Y' in row:
        return 'Model Y'
    elif 'Model 3' in row:
        return 'Model 3'


    
df_2['Modelo'] = df_2['Modelo'].apply(lambda x : clean_model(x))
df_1['model'] = df_1['model'].apply(lambda x : clean_model(x))
df_2
```

```python
df_2.rename(columns={'Modelo' : 'model', 'Precio (€)' : 'price', 'Año' : 'year', 'Kilometraje (kms)' : 'km', 'Color' : 'color', 'País' : 'country'}, inplace=True)
df_2['price'] = df_2['price'].astype(str).apply(lambda row: row.replace('.', '').replace(',', '')).astype(int)
df_2['km'] = df_2['km'].astype(str).apply(lambda row: row.replace('.', '').replace(',', '')).astype(int)
df_2
```

```python
df_1 = df_1[['model', 'price', 'year', 'km', 'color', 'País']]
df_1.rename(columns={'Modelo' : 'model', 'Precio (€)' : 'price', 'Año' : 'year', 'Kilometraje (kms)' : 'km', 'Color' : 'color', 'País' : 'country'}, inplace=True)
df_1
```

```python
df_1['km'] = df_1['km'].astype(str).apply(lambda row: row.replace('.', '').replace(',', '')).astype(int)
df_1
```

```python
df = pd.concat([df_1, df_2])
df
```

```python
df = pd.concat([df_1, df_2])
rows_to_drop = df.isin(['Desconocido']).any(axis=1)
df = df[~rows_to_drop]
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```
