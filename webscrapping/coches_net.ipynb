```python
import pandas as pd
import numpy as np
import shutil
import pyautogui
import time

import requests
from bs4 import BeautifulSoup

df = pd.DataFrame(columns=['model', 'color', 'name','year', 'km', 'price'])
```

```python
urls = {
    'Model S' : 1171,
    'Model 3' : 1262,
    'Model X' : 1172,
    'Model Y' : 1328
}
colors = {

    #1 : 'Amarillo',
    #2 : 'Azul',
    #3 : 'Beige',
    4 : 'Blanco',
    5 : 'Granate',
    6 : 'Gris/Plata',
    #7 : 'Marron',
    #8 : 'Naranja',
    9 : 'Negro',
    10 : 'Rojo',
    #11 : 'Rosa',
    #12 : 'Verde',
    #13 : 'Violeta/Lila'
}
```

```python
import os

# Buscar el único archivo .html en el directorio actual
```

```python
def get_car_info(soup, model_name, color):
    try:
        section = soup.find('section', class_='mt-AdsList-content')
        coches = [div for div in soup.find_all('div') if div.has_attr('data-ad-id')]
        for coche in coches:
            model = coche.find('h2').text
            attrs = [i.text for i in coche.find('ul', class_='mt-CardAd-attr').find_all('li')]
            year = int(list(filter(lambda x: x.isnumeric() and len(str(x)) == 4, attrs))[0])
            km = list(filter(lambda x : 'km' in x, attrs))[0].split()[0]
            price = coche.find('h3').text
            df.loc[len(df)] = [model_name, color, model, year, km, price]
            print('LONGITUD: ',len(df))
    except:
        return None
        
```

```python
time.sleep(2)
for model, model_code in urls.items():
    for color_code, color_name in colors.items():
        url = f"https://www.coches.net/search/?ColourId={color_code}&MakeIds%5B0%5D=1354&ModelIds%5B0%5D={model_code}"
        print(model, color_name)
        print('URL: ',url)
        pyautogui.moveTo(400, 100)  # Ejemplo: posición del buscador
        pyautogui.click()  # Clic en la barra de búsqueda

        for char in url:
            if char == '/':
                pyautogui.hotkey('shift', '7')  # Barra inclinada
            elif char == ':':
                pyautogui.hotkey('shift', '.')  # Dos puntos
            elif char == '=':
                pyautogui.hotkey('shift', '0')  # Igual
            elif char == '?':
                pyautogui.hotkey('shift', "'")  # Signo de interrogación
            elif char == '%':
                pyautogui.hotkey('shift', '5')  # Porcentaje
            elif char == '[':
                pyautogui.hotkey('altgr', '`')  # Corchete abierto
            elif char == ']':
                pyautogui.hotkey('altgr', '+')  # Corchete cerrado
            else:
                pyautogui.write(char, interval=0)  # Escribir caracteres normales

        pyautogui.press('enter')
        pyautogui.moveTo(400, 300)
        for _ in range(10):  # Ajusta el número de iteraciones según la longitud de la página
            pyautogui.scroll(-500)  # Desplazamiento hacia abajo
            time.sleep(1)  # Pausa de 1 segundo entre desplazamientos

        print("Scroll hasta abajo completado.")
        pyautogui.hotkey('ctrl', 's')
        time.sleep(2) 
        pyautogui.moveTo(1850, 380)
        pyautogui.click()
        time.sleep(6)
        current_dir = os.getcwd()
        html_file = None

        # Iterar para encontrar el archivo .html
        for file in os.listdir(current_dir):
            if file.endswith(".html"):
                html_file = file
                break

        if html_file:
            print(f"Archivo encontrado: {html_file}")
            assets_dir = html_file.replace(".html", "_files")

            # Abrir el archivo y convertirlo en un BeautifulSoup
            with open(html_file, "r", encoding="utf-8") as file:
                soup = BeautifulSoup(file, "html.parser")
                get_car_info(soup, model, color_name)

            # Eliminar el archivo HTML y la carpeta de assets
            os.remove(html_file)
            print(f"Archivo HTML eliminado: {html_file}")

            if os.path.exists(assets_dir):
                shutil.rmtree(assets_dir)
                print(f"Directorio eliminado: {assets_dir}")
        else:
            print("No se encontró ningún archivo HTML guardado.")
        print(f"{model} {color_name} Scrapeado!")
```
